```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('2_sorting.csv')
print(df.to_string(index=False))
```

```python
# Chain Graph: Sequential vs Parallel time across input sizes
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Assignment 2 - Sorting Chain Graph', fontsize=14)

# Bubble Sort chain graph
axes[0].plot(df['size'], df['seq_bubble'], 'o-', label='Sequential Bubble', color='blue')
axes[0].plot(df['size'], df['par_bubble'], 's--', label='Parallel Bubble (Odd-Even)', color='red')
axes[0].set_title('Bubble Sort: Sequential vs Parallel Time')
axes[0].set_xlabel('Input Size')
axes[0].set_ylabel('Time (seconds)')
axes[0].legend()
axes[0].grid(True)

# Merge Sort chain graph
axes[1].plot(df['size'], df['seq_merge'], 'o-', label='Sequential Merge', color='blue')
axes[1].plot(df['size'], df['par_merge'], 's--', label='Parallel Merge', color='red')
axes[1].set_title('Merge Sort: Sequential vs Parallel Time')
axes[1].set_xlabel('Input Size')
axes[1].set_ylabel('Time (seconds)')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.savefig('2_chain_graph.png', dpi=150)
plt.show()
```

```python
# Speedup and Efficiency Graph
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Assignment 2 - Sorting Speedup & Efficiency', fontsize=14)

axes[0].plot(df['size'], df['speedup_bubble'], 'o-', label='Speedup Bubble', color='green')
axes[0].plot(df['size'], df['speedup_merge'],  's--', label='Speedup Merge',  color='orange')
axes[0].axhline(y=1, color='gray', linestyle=':', label='Speedup = 1')
axes[0].set_title('Speedup vs Input Size')
axes[0].set_xlabel('Input Size')
axes[0].set_ylabel('Speedup')
axes[0].legend()
axes[0].grid(True)

axes[1].plot(df['size'], df['efficiency'], 'D-', label='Efficiency', color='purple')
axes[1].set_title('Efficiency vs Input Size')
axes[1].set_xlabel('Input Size')
axes[1].set_ylabel('Efficiency')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.savefig('2_speedup_efficiency.png', dpi=150)
plt.show()
```
