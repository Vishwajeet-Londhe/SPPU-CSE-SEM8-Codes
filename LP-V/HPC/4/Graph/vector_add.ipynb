```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data (ensure `vector_add.csv` is in the same folder as this notebook)
df = pd.read_csv('vector_add.csv')
print(df.to_string(index=False))
```

```python
# Chain Graph: Sequential vs Parallel time across input sizes
fig, ax = plt.subplots(figsize=(10, 5))
fig.suptitle('Vector Addition: Sequential vs Parallel Time', fontsize=14)

ax.plot(df['size'], df['seq_time'], 'o-', label='CPU Sequential', color='blue')
ax.plot(df['size'], df['par_time'], 's--', label='GPU Parallel', color='red')
ax.set_xscale('log')
ax.set_title('Chain Graph')
ax.set_xlabel('Input Size')
ax.set_ylabel('Time (seconds)')
ax.legend()
ax.grid(True, which='both', linestyle=':' )

plt.tight_layout()
plt.savefig('vector_chain_graph.png', dpi=150)
plt.show()
```

```python
# Speedup and Efficiency Graph
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Vector Addition: Speedup & Efficiency', fontsize=14)

axes[0].plot(df['size'], df['speedup'], 'o-', label='Speedup', color='green')
axes[0].axhline(y=1, color='gray', linestyle=':', label='Speedup = 1')
axes[0].set_xscale('log')
axes[0].set_title('Speedup vs Input Size')
axes[0].set_xlabel('Input Size')
axes[0].set_ylabel('Speedup')
axes[0].legend()
axes[0].grid(True, which='both', linestyle=':' )

axes[1].plot(df['size'], df['efficiency'], 'D-', label='Efficiency', color='purple')
axes[1].set_xscale('log')
axes[1].set_title('Efficiency vs Input Size')
axes[1].set_xlabel('Input Size')
axes[1].set_ylabel('Efficiency')
axes[1].legend()
axes[1].grid(True, which='both', linestyle=':' )

plt.tight_layout()
plt.savefig('vector_speedup_efficiency.png', dpi=150)
plt.show()
```
