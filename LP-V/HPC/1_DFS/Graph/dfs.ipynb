```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dfs.csv')
print(df.to_string(index=False))
```

```python
# DFS Chain Graph: Sequential vs Parallel time across input sizes
fig, ax = plt.subplots(figsize=(8, 5))
fig.suptitle('Assignment 1 - DFS Chain Graph', fontsize=14)

ax.plot(df['size'], df['seqdfs'], 'o-', label='Sequential DFS', color='blue')
ax.plot(df['size'], df['pardfs'], 's--', label='Parallel DFS', color='red')
ax.set_title('DFS: Sequential vs Parallel Time')
ax.set_xlabel('Input Size (vertices)')
ax.set_ylabel('Time (seconds)')
ax.legend()
ax.grid(True)

plt.tight_layout()
plt.savefig('1_chain_graph_dfs.png', dpi=150)
plt.show()
```

```python
# DFS Speedup and Efficiency Graph
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Assignment 1 - DFS Speedup & Efficiency', fontsize=14)

axes[0].plot(df['size'], df['speedup'], 'o-', label='Speedup DFS', color='orange')
axes[0].axhline(y=1, color='gray', linestyle=':', label='Speedup = 1')
axes[0].set_title('Speedup vs Input Size')
axes[0].set_xlabel('Input Size (vertices)')
axes[0].set_ylabel('Speedup')
axes[0].legend()
axes[0].grid(True)

axes[1].plot(df['size'], df['efficiency'], 'D-', label='Efficiency', color='purple')
axes[1].set_title('Efficiency vs Input Size')
axes[1].set_xlabel('Input Size (vertices)')
axes[1].set_ylabel('Efficiency')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.savefig('1_speedup_efficiency_dfs.png', dpi=150)
plt.show()
```
