```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('3_reduction.csv')
print(df.to_string(index=False))
```

```python
# Reduction Chain Graph: Sequential vs Parallel time across input sizes
fig, ax = plt.subplots(figsize=(8, 5))
fig.suptitle('Assignment 3 - Reduction Chain Graph', fontsize=14)

ax.plot(df['size'], df['seq_time'], 'o-', label='Sequential', color='blue')
ax.plot(df['size'], df['par_time'], 's--', label='Parallel', color='red')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_title('Reduction: Sequential vs Parallel Time')
ax.set_xlabel('Input Size')
ax.set_ylabel('Time (seconds)')
ax.legend()
ax.grid(True, which='both', linestyle=':' )

plt.tight_layout()
plt.savefig('3_chain_graph_reduction.png', dpi=150)
plt.show()
```

```python
# Speedup and Efficiency Graph
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Reduction Speedup & Efficiency', fontsize=14)

axes[0].plot(df['size'], df['speedup'], 'o-', label='Speedup', color='green')
axes[0].axhline(y=1, color='gray', linestyle=':', label='Speedup = 1')
axes[0].set_xscale('log')
axes[0].set_title('Speedup vs Input Size')
axes[0].set_xlabel('Input Size')
axes[0].set_ylabel('Speedup')
axes[0].legend()
axes[0].grid(True, which='both', linestyle=':' )

axes[1].plot(df['size'], df['efficiency'], 'D-', label='Efficiency', color='purple')
axes[1].set_xscale('log')
axes[1].set_title('Efficiency vs Input Size')
axes[1].set_xlabel('Input Size')
axes[1].set_ylabel('Efficiency')
axes[1].legend()
axes[1].grid(True, which='both', linestyle=':' )

plt.tight_layout()
plt.savefig('3_speedup_efficiency_reduction.png', dpi=150)
plt.show()
```
